```python
import pandas as pd
train_df = pd.read_csv('/kaggle/input/problem-3-republic-olymp-tst-homework/train.csv')
```

```python
train_df
```

```python
train_df['label'] = 1.0
```

```python
train_df
```

```python
import numpy as np
```

```python
for i in range(len(train_df)):
    kol = 0
    while(kol<=1):
        val = np.random.randint(len(train_df))
        if val != i:
            kol+=1
            train_df.loc[len(train_df)] = [train_df.loc[i,'description'],train_df.loc[val,'conclusion'],0]
```

```python
train_df
```

```python
import re

def clean_text(text):
    text = text.lower()  # lowercase
    text = re.sub(r'\s+', ' ', text).strip()  # remove extra spaces
    return text

# Применяем очистку
train_df['description'] = train_df['description'].apply(clean_text)
train_df['conclusion'] = train_df['conclusion'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

all_texts = train_df['description'].tolist() + train_df['conclusion'].tolist()

tfidf = TfidfVectorizer(max_features=50000)  # можно увеличить/уменьшить
tfidf.fit(all_texts)

desc_tfidf = tfidf.transform(train_df['description'].tolist()).toarray()
conc_tfidf = tfidf.transform(train_df['conclusion'].tolist()).toarray()

X = np.hstack([desc_tfidf, conc_tfidf, np.abs(desc_tfidf - conc_tfidf)])
y = train_df['label'].values
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, random_state=42)

model = CatBoostClassifier(
)
model.fit(X_train, y_train)

y_pred = model.predict_proba(X_val)[:, 1]
auc = roc_auc_score(y_val, y_pred)
print(f"AUC: {auc:.4f}")
```

```python
test_batches = pd.read_csv('/kaggle/input/problem-3-republic-olymp-tst-homework/test_batches.csv')
test_batches
```

```python
test_conclusions = pd.read_csv('/kaggle/input/problem-3-republic-olymp-tst-homework/test_conclusions.csv')
test_conclusions
```

```python
test_descriptions = pd.read_csv('/kaggle/input/problem-3-republic-olymp-tst-homework/test_descriptions.csv')
```

```python
test_descriptions
```

```python
s1 = []
s2 = []
```

```python
d1 = {}
for i in range(len(test_descriptions)):
    d1[test_descriptions.loc[i,'description_id']] = test_descriptions.loc[i,'description']
```

```python
d2 = {}
for i in range(len(test_conclusions)):
    d2[test_conclusions.loc[i,'conclusion_id']] = test_conclusions.loc[i,'conclusion']
```

```python
# Предполагается: d1 и d2 — словари {id: текст}
# Очистим их тоже
d1 = {k: clean_text(v) for k, v in d1.items()}
d2 = {k: clean_text(v) for k, v in d2.items()}
```

```python
import ast
```

```python
for i in range(len(test_batches)):
    print(i)
    for j in ast.literal_eval(test_batches.loc[i,'description_ids']):
        s1.append(j)
        mx = 0
        id1 = ast.literal_eval(test_batches.loc[i,'conclusion_ids'])[0]
        for k in ast.literal_eval(test_batches.loc[i,'conclusion_ids']):
            desc_text = d1[j]
            concl_text = d2[k]
            desc_vec = tfidf.transform([desc_text]).toarray()[0]
            concl_vec = tfidf.transform([concl_text]).toarray()[0]
            x = np.hstack([desc_vec, concl_vec, np.abs(desc_vec - concl_vec)])
            pred_prob = model.predict_proba([x])[0][1]
            if pred_prob > mx:
                mx = pred_prob
                id1 = k
        s2.append(id1)
```
